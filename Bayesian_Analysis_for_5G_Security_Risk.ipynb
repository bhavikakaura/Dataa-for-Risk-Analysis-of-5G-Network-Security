```python
import pandas as pd
import io

# Load uploaded CSV file
df = pd.read_csv(io.BytesIO(uploaded['realistic_5G_security_dataset (1).csv']))

# Define conditions
high_impact = df['Economic_Impact_kCAD'] > 500
high_cvss = df['CVSS_Score'] > 8
ddos_attack = df['Threat_Type'] == 'DDoS'

# Prior: P(High Impact)
P_high_impact = high_impact.mean()

# Likelihood: P(High CVSS | High Impact)
P_high_cvss_given_high_impact = df[high_impact]['CVSS_Score'].gt(8).mean()

# Marginal: P(High CVSS)
P_high_cvss = high_cvss.mean()

# Posterior: P(High Impact | High CVSS) using Bayes' Theorem
P_high_impact_given_high_cvss = (P_high_cvss_given_high_impact * P_high_impact) / P_high_cvss

# Threat Origin Example: Bayes on DDoS detection
P_ddos = ddos_attack.mean()
P_detected_given_ddos = 0.80  # Assumed
P_detected = 0.54  # Assumed

# P(DDoS | Detected)
P_ddos_given_detected = (P_detected_given_ddos * P_ddos) / P_detected

# Output
print(f"Prior P(High Impact): {P_high_impact:.3f}")
print(f"Likelihood P(High CVSS | High Impact): {P_high_cvss_given_high_impact:.3f}")
print(f"Posterior P(High Impact | High CVSS): {P_high_impact_given_high_cvss:.3f}")
print(f"Bayes: P(DDoS | Detected Breach): {P_ddos_given_detected:.3f}")
```
